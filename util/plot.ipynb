```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../output/marl/ppo-2000s-2/colombo_conn0_ep62.csv')
```

```python
x = df['step']
y1 = df['system_mean_waiting_time_mc']
y2 = df['system_mean_waiting_time_mb']
```

```python
plt.plot(x, y1, label='system_mean_waiting_time_mc')
plt.plot(x, y2, label='system_mean_waiting_time_mb')

plt.title('Line Graph of Y1 and Y2 Columns')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend()
```

```python
mc_mean = y1.mean()
mb_mean = y2.mean()

print('Mean of other vehicles: ', mc_mean)
print('Mean of two wheelers', mb_mean)
```
