```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_marl = pd.read_csv('../output/custom-eval/ppo-3000s-2_conn0_ep2.csv')
df_fixed = pd.read_csv('../output/fixed_time_phase/3000s1.csv')
```

```python
x = df_fixed['step']
y1 = df_fixed['system_mean_waiting_time_mc']
y2 = df_fixed['system_mean_waiting_time_mb']

x_m = df_marl['step']
y1_m = df_marl['system_mean_waiting_time_mc']
y2_m = df_marl['system_mean_waiting_time_mb']
```

```python
plt.plot(x_m, y1_m, label='Normal Vehicles')
plt.plot(x_m, y2_m, label='Two Wheelers')

plt.title('Line Graph of Mean Waiting Time of Normal Vehicles and Two Whelers (MARL)')
plt.xlabel('Step')
plt.ylabel('Mean Waiting Time')
plt.legend()
```

```python
plt.plot(x, y1, label='Normal Vehicles')
plt.plot(x, y2, label='Two Wheelers')

plt.title('Line Graph of Mean Waiting Time of Normal Vehicles and Two Whelers')
plt.xlabel('Step')
plt.ylabel('Mean Waiting Time')
plt.legend()
```

```python
mc_mean_marl = y1_m.mean()
mb_mean_marl = y2_m.mean()

mc_mean = y1.mean()
mb_mean = y2.mean()

print('Mean of other vehicles (MARL): ', mc_mean_marl)
print('Mean of two wheelers: (MARL)', mb_mean_marl)

print('Mean of other vehicles: ', mc_mean)
print('Mean of two wheelers: ', mb_mean)
```
