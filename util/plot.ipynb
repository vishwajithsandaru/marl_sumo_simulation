```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_marl = pd.read_csv('../output/marl/ppo-2000s-2/colombo_conn0_ep62.csv')
df_fixed = pd.read_csv('../output/fixed_time_phase/5000.csv')
```

```python
x = df_fixed['step']
y1 = df_fixed['system_mean_waiting_time_mc']
y2 = df_fixed['system_mean_waiting_time_mb']
```

```python
plt.plot(x, y1, label='Normal Vehicles')
plt.plot(x, y2, label='Two Wheelers')

plt.title('Line Graph of Mean Waiting Time of Normal Vehicles and Two Whelers')
plt.xlabel('Step')
plt.ylabel('Mean Waiting Time')
plt.legend()
```

```python
mc_mean = y1.mean()
mb_mean = y2.mean()

print('Mean of other vehicles: ', mc_mean)
print('Mean of two wheelers', mb_mean)
```
